```python
import pandas as pd
```

```python
df = pd.read_csv("books.csv")
```

```python
df
```

```python
def task2():
    df["vowels"] = df["Author"].str.split(", ", expand=True)[1].str.count(r"[AEIOUaeiou]").fillna(0)
task2()
df = df.astype({'vowels': 'int32'})
df
```

```python
def task3():
    return df[df["Author"].str.split(", ", expand=True)[0].str.contains(r"[AEIOUaeiou][AEIOUaeiou][AEIOUaeiou]", na=False)]
task3()
```

```python
def task4():
    return df[df["Title"].str.count(r"[, ]") == 0]
task4()
```

```python
def task5():
    df["Number in title"] = df.apply(lambda x: 1 if  any(map(str.isdigit, x["Title"])) else 0, axis=1)
    return df["Number in title"]
task5()
df[df["Title"].str.contains(r"[0123456789]")]
```

```python
df
```

```python
def task6():
    df["Last word in book title"] = df.apply(lambda x: x["Title"].split(" ")[len(x["Title"].split(" ")) - 1], axis=1)
    return df["Last word in book title"]
task6()
```

```python
def task7():
    return df[df["Publisher"].str.count(r"[ ]") == 1]
task7()
```

```python
def task8():
    return df[df["Author"].str.split(", ", expand=True)[0].str.match(r"..[y].*", na=False)]
task8()
```
