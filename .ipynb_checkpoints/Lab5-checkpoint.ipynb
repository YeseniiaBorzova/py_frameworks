```python
import pandas as pd
import numpy as np

data = pd.read_excel("En_In.xls")
df = data.copy()
df = df.drop(columns=["Unnamed: 0","Unnamed: 1" ])
df = df.rename(columns={"Unnamed: 2":"Country","Unnamed: 3":"Energy Supply","Unnamed: 4":"Energy Supply per Capita","Unnamed: 5":"% Renewable"})
df = df.drop(df.index[0:17])
df = df.drop(df.index[227:282])
df = df.replace('...', np.NaN)
df = df.reset_index(drop=True)
df["Energy Supply"] = df["Energy Supply"]*1000000
for index in df.index:
    if df.loc[index,'Country']=="Republic of Korea":
        df.loc[index,'Country'] = "South Korea"
    elif df.loc[index,'Country']=="United States of America20":
        df.loc[index,'Country'] = "United States"
    elif df.loc[index,'Country']=="United Kingdom of Great Britain and Northern Ireland19":
        df.loc[index,'Country'] = "United Kingdom"
    elif df.loc[index,'Country']=="China, Hong Kong Special Administrative Region3":
        df.loc[index,'Country'] = "Hong Kong"
```

```python
def find_between( s, first, last ):
    try:
        start = s.index( first ) + len( first )
        end = s.index( last, start )
        return s[start:end]
    except ValueError:
        return ""

for index in df.index:
    for i in df.loc[index,'Country']:
        if i.isdigit():
            df.loc[index,'Country'] = df.loc[index,'Country'].replace(i,"")

    
for index in df.index:
    start = " ("
    end = ")"
    substring = find_between(df.loc[index,'Country'],start, end)
    df.loc[index,'Country'] = df.loc[index,'Country'].replace(start+substring+end,'')
```

```python
gpd = pd.read_csv('gpd.csv',skiprows=4)
gpd.loc[gpd.loc[gpd['Country Name']=="Korea, Rep."].index[0],'Country Name'] = "South Korea"
gpd.loc[gpd.loc[gpd['Country Name']=="Iran, Islamic Rep."].index[0],'Country Name'] = "Iran"
gpd.loc[gpd.loc[gpd['Country Name']=="Hong Kong SAR, China"].index[0],'Country Name'] = "Hong Kong"
gpd = gpd.rename(columns={'Country Name':'Country'})
gpd.head(1)
```

```python
scim = pd.read_excel("scimagojr.xlsx")
scim
```

```python
cut_gpd = gpd[['Country','Country Code','Indicator Name', 'Indicator Code','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']].copy()
cut_scim = scim.head(15)
intersect_df = pd.merge(pd.merge(df, cut_gpd, on="Country"),cut_scim,on='Country')
intersect_df = intersect_df.drop(columns=['Country Code','Indicator Name','Indicator Code'])
intersect_df = intersect_df[['Country','Rank','Documents','Citable documents','Citations','Self-citations','Citations per document','H index','Energy Supply','Energy Supply per Capita','% Renewable','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']]    
intersect_df = intersect_df.reset_index(drop=True)
intersect_df = intersect_df.set_index('Country')
intersect_df.sort_values('Rank', ascending=True).head(3)
```

```python
intersect_df.shape
```

```python
years = ['2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']
def task_8():
    df = intersect_df[years].mean(axis=1,skipna=True)
    return df.sort_values(ascending=False)
print(task_8())
```

```python
years = ['2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']
def task_9():
    country = task_8()
    name = country.index[4]
    return (name, intersect_df.loc[name]['2015']-intersect_df.loc[name]['2006'])

print(task_9())
```

```python
def task_10():
    max_country = intersect_df.groupby(["% Renewable"])[['Rank']].max().index[-1]
    name = intersect_df.loc[intersect_df['% Renewable']==max_country].index[0]
    return (name,max_country)
print(task_10())
```

```python
intersect_df["Population"] = intersect_df["Energy Supply"] / intersect_df["Energy Supply per Capita"]
def task_11():
    pop = intersect_df.groupby(["Population"])[['Rank']].max()
    name = intersect_df.loc[intersect_df['Population']==pop.index[-6]].index[0]
    return (name, pop.index[-6])
print(task_11())   
```

```python
intersect_df["Docs per person"] = intersect_df['Citable documents'] / intersect_df['Population']
def task_12():
    col_1 = intersect_df["Docs per person"]
    col_2 = intersect_df["Energy Supply per Capita"]
    return col_1.corr(col_2)
print(task_12())
```

```python
median = intersect_df["% Renewable"].describe().values[5]
#intersect_df["% Renewable"].median()
intersect_df = intersect_df.sort_values('Rank',ascending=True)
def task_13():
    intersect_df["Above median"] = intersect_df["% Renewable"].apply(lambda x: 1 if x>=median else 0 )
    return intersect_df["Above median"]
print(task_13())
```

```python
ContinentDict = {'China':'Asia',
'United States':'North America',
'Japan':'Asia',
'United Kingdom':'Europe',
'Russian Federation':'Europe',
'Canada':'North America',
'Germany':'Europe',
'India':'Asia',
'France':'Europe',
'South Korea':'Asia',
'Italy':'Europe',
'Spain':'Europe',
'Iran':'Asia',
'Australia':'Australia',
'Brazil':'South America'}
# def task14():
#     continets = pd.DataFrame(columns = ['size', 'sum', 'mean', 'std'])
#     for i, j in intersect_df.groupby(ContinentDict):
#         continets.loc[i] = [len(j), j['Population'].sum(),j['Population'].mean(),j['Population'].std()]
#     return continets
# task14()
```

```python
def task_14():
    as_count, aus_count, eu_count, na_count, sa_count = 0,0,0,0,0
    as_list,aus_list,eu_list,na_list,sa_list=[],[],[],[],[]
    continents = pd.DataFrame(columns = ['Continent','size','sum','mean','std'])
    #continents = continents.set_index('Continent')
    amount_dic={}
    con_dic={}
    for index in intersect_df.index:
        for key in ContinentDict:
            if key==index and ContinentDict[key]=='North America':
                na_name = 'North America'
                na_count+=1
                na_list.append(key)
            elif key==index and ContinentDict[key]=='Asia':
                as_name = 'Asia'
                as_count+=1
                as_list.append(key)
            elif key==index and ContinentDict[key]=='Australia':
                aus_name = 'Australia'
                aus_count+=1
                aus_list.append(key)
            elif key==index and ContinentDict[key]=='Europe':
                eu_name = 'Europe'
                eu_count+=1
                eu_list.append(key)
            elif key==index and ContinentDict[key]=='South America':
                sa_name = 'South America'
                sa_count+=1
                sa_list.append(key)
    amount_dic[na_name]=na_count 
    amount_dic[as_name]=as_count 
    amount_dic[aus_name]=aus_count 
    amount_dic[eu_name]=eu_count 
    amount_dic[sa_name]=sa_count 
    con_dic[na_name]=na_list
    con_dic[as_name]=as_list
    con_dic[aus_name]=aus_list
    con_dic[eu_name]=eu_list
    con_dic[sa_name]=sa_list
    as_sum=intersect_df[(intersect_df.index=='China') | (intersect_df.index=='India') |(intersect_df.index=='Iran')|(intersect_df.index=='Japan')|(intersect_df.index=='South Korea')]['Population'].sum()
    na_sum=intersect_df[(intersect_df.index=='Canada') | (intersect_df.index=='United States')]['Population'].sum()
    aus_sum=intersect_df[intersect_df.index=='Australia']['Population'].sum()
    eu_sum=intersect_df[(intersect_df.index=='France') | (intersect_df.index=='Germany') |(intersect_df.index=='Italy')|(intersect_df.index=='Russian Federation')|(intersect_df.index=='Spain')|(intersect_df.index=='United Kingdom')]['Population'].sum()
    sa_sum=intersect_df[intersect_df.index=='Brazil']['Population'].sum()
    as_mean=intersect_df[(intersect_df.index=='China') | (intersect_df.index=='India') |(intersect_df.index=='Iran')|(intersect_df.index=='Japan')|(intersect_df.index=='South Korea')]['Population'].mean()
    na_mean=intersect_df[(intersect_df.index=='Canada') | (intersect_df.index=='United States')]['Population'].mean()
    aus_mean=intersect_df[intersect_df.index=='Australia']['Population'].mean()
    eu_mean=intersect_df[(intersect_df.index=='France') | (intersect_df.index=='Germany') |(intersect_df.index=='Italy')|(intersect_df.index=='Russian Federation')|(intersect_df.index=='Spain')|(intersect_df.index=='United Kingdom')]['Population'].mean()
    sa_mean=intersect_df[intersect_df.index=='Brazil']['Population'].mean()
    as_std=intersect_df[(intersect_df.index=='China') | (intersect_df.index=='India') |(intersect_df.index=='Iran')|(intersect_df.index=='Japan')|(intersect_df.index=='South Korea')]['Population'].std()
    na_std=intersect_df[(intersect_df.index=='Canada') | (intersect_df.index=='United States')]['Population'].std()
    aus_std=intersect_df[intersect_df.index=='Australia']['Population'].std()
    eu_std=intersect_df[(intersect_df.index=='France') | (intersect_df.index=='Germany') |(intersect_df.index=='Italy')|(intersect_df.index=='Russian Federation')|(intersect_df.index=='Spain')|(intersect_df.index=='United Kingdom')]['Population'].std()
    sa_std=intersect_df[intersect_df.index=='Brazil']['Population'].std()
    continents["Continent"]=amount_dic.keys()
    continents["size"]=amount_dic.values()
    continents['sum']=[na_sum, as_sum, aus_sum, eu_sum,sa_sum]
    continents['mean']=[na_mean, as_mean, aus_mean, eu_mean,sa_mean]
    continents['std']=[na_std, as_std, aus_std, eu_std,sa_std]
    continents = continents.reset_index(drop=True)
    continents = continents.set_index('Continent')
    return continents.sort_values('Continent', ascending=True)
```
