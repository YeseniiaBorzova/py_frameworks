```python
import pandas as pd
import numpy as np

df = pd.read_csv('books.csv.csv')
df.head(20)
```

```python
df["Number of vowels"] = df["Author"].str.split(", ", expand= True)[1].str.count(r'[AEIOUaeiou]').fillna(0)
df.head(5)
```

```python
df[df["Author"].str.split(", ", expand= True)[0].str.contains(r'[AEIOUaeiou]{3}', na=False)]
```

```python
df[df['Title'].str.contains(r'\b\w{1}\b').fillna(0)]
```

```python
df['Contains digits'] = df.apply(lambda x: 1 if any(map(str.isdigit, x['Title'])) else 0, axis=1)
df[df['Title'].str.contains(r'[0-9]').fillna(0)]
```

```python
df['Last word'] = df.apply(lambda x: x['Title'].split(" ")[len(x["Title"].split(" "))-1], axis=1)
df.head(5)
```

```python
df[df['Publisher'].str.count(r"[ ]")==1]
```

```python
df[df["Author"].str.split(", ", expand= True)[0].str.match(r'..[y].*',na=False)]
```

```python
df[df["Author"].str.split(", ", expand=True)[1].str.count(r".") < df["Author"].str.split(", ", expand=True)[0].str.count(r".")]
#surname>name
```
